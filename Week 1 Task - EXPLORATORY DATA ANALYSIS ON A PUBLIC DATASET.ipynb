# EXPLORATORY DATA ANALYSIS (Titanic Dataset)

## Step 1: Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load the Dataset

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

### Display first few rows

```python
print("First 5 rows of the dataset:")
print(df.head())
```

## Step 3: Understand the Dataset Structure

```python
print("\nDataset Information:")
print(df.info())
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

## Step 4: Check for Missing and Duplicate Values

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDuplicate Records:", df.duplicated().sum())
```

## Step 5: Data Preprocessing

### Fill missing Age values with median

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

### Fill missing Embarked values with mode

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

### Drop duplicate rows

```python
df.drop_duplicates(inplace=True)
```

### Drop Cabin Column

```python
df.drop('Cabin', axis=1, inplace=True)
```

### Verify missing values after cleaning

```python
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

## Step 6: Visualization

### Survival count

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

### Age distribution

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Survival by Gender

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

### Survival by Passenger Class

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

## Step 9: Key Insight

```python
print("\nKey Insights:")
print("- Females had a higher survival rate than males.")
print("- Passengers in 1st class survived more compared to 2nd and 3rd class.")
print("- Younger passengers had slightly higher survival chances.")
print("- Passenger class and fare show correlation with survival.")
```
